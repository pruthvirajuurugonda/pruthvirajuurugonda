```python
#Python Program for Prime Number

num = int(input("Enter number = "))

prime = (num%2)
if prime==0:
  print("given number is not prime:", num)
else:
  print("Given number is prime:", num)
```

```python
#Prime numbers in give range
STR_num = int(input("Enter start number = "))
END_num = int(input("Enter end number = "))

#def prime(x, y):
prime_list = []
for i in range(STR_num, END_num):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                print("I value:",i)
                print("J value:",j)
                if i % j == 0:
                    break
            else:
                prime_list.append(i) 
                        
print(prime_list)
```
